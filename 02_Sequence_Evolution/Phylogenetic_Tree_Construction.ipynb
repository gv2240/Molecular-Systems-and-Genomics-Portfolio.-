```python
from Bio.Align.Applications import ClustalwCommandline, ClustalOmegaCommandline
from Bio import AlignIO
from Bio.Phylo.TreeConstruction import DistanceCalculator, DistanceTreeConstructor
from Bio import Phylo
import matplotlib.pyplot as plt

# --- Run ClustalW (requires clustalw binary installed in environment) ---
# clustalw_cline = ClustalwCommandline("clustalw", infile="sequences.fasta")
# stdout, stderr = clustalw_cline()

# --- Run Clustal Omega (requires clustalo binary installed in environment) ---
clustalomega_cline = ClustalOmegaCommandline(infile="/content/sequence.fasta",
                                             outfile="/content/aligned.fasta",
                                             verbose=True,
                                             auto=True,
                                             force=True)
stdout, stderr = clustalomega_cline()

# --- Load alignment ---
alignment = AlignIO.read("aligned.fasta", "fasta")
print(alignment)

# --- Visualize alignment ---
fig, ax = plt.subplots(figsize=(12, 6))
ax.imshow([[ord(c) for c in record.seq] for record in alignment],
          aspect="auto", cmap="viridis")
ax.set_title("Clustal Alignment")
plt.show()

# --- Build phylogenetic tree ---
calculator = DistanceCalculator('identity')
dm = calculator.get_distance(alignment)

constructor = DistanceTreeConstructor(calculator, 'nj')
tree = constructor.build_tree(alignment)

Phylo.draw(tree)
```
