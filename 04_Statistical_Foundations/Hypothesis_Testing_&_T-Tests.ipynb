```python
import numpy as np

# Define parameters for the population
n_population = 10000
population_mean = 10
population_std = 1 # Choosing a standard deviation of 1 for demonstration

# Generate a random population dataset with a normal distribution
population = np.random.normal(loc=population_mean, scale=population_std, size=n_population)

print(f"Generated population size: {len(population)}")
print(f"Mean of the generated population: {np.mean(population):.4f}")
print(f"Standard deviation of the generated population: {np.std(population):.4f}")

# Display the first 10 elements to verify
print("First 10 elements of the population:")
print(population[:10])
```

```python
import numpy as np

# Take a random sample of 100 from the population
sample_size = 100
sample = np.random.choice(population, size=sample_size, replace=False)

# Calculate mean of the sample
sample_mean = np.mean(sample)

# Calculate standard deviation of the sample
sample_std = np.std(sample, ddof=1) # ddof=1 for sample standard deviation

# Calculate standard error of the mean (SEM)
# SEM = sample_std / sqrt(sample_size)
std_error_mean = sample_std / np.sqrt(sample_size)

print(f"Sample size: {len(sample)}")
print(f"Sample mean: {sample_mean:.4f}")
print(f"Sample standard deviation: {sample_std:.4f}")
print(f"Standard error of the mean: {std_error_mean:.4f}")

print("\nFirst 10 elements of the sample:")
print(sample[:10])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot histogram for the population
sns.histplot(population, kde=True, ax=axes[0], color='skyblue')
axes[0].set_title('Distribution of Population (n=10000)')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')
axes[0].axvline(np.mean(population), color='red', linestyle='dashed', linewidth=1, label=f'Mean: {np.mean(population):.2f}')
axes[0].legend()

# Plot histogram for the sample
sns.histplot(sample, kde=True, ax=axes[1], color='lightcoral')
axes[1].set_title('Distribution of Sample (n=100)')
axes[1].set_xlabel('Value')
axes[1].set_ylabel('Frequency')
axes[1].axvline(np.mean(sample), color='darkblue', linestyle='dashed', linewidth=1, label=f'Mean: {np.mean(sample):.2f}')
axes[1].legend()

plt.tight_layout()
plt.show()
```

```python
from scipy import stats
import numpy as np

# The 'sample' and 'population_mean' variables are available from previous cells.

# Perform a one-sample t-test
# We are testing if the mean of our 'sample' is significantly different from the 'population_mean'
t_statistic, p_value = stats.ttest_1samp(sample, population_mean)

print(f"Sample mean: {np.mean(sample):.4f}")
print(f"Hypothesized population mean: {population_mean}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results (common alpha level is 0.05)
alpha = 0.05
if p_value < alpha:
    print(f"\nSince p-value ({p_value:.4f}) < alpha ({alpha}), we reject the null hypothesis.")
    print("There is a statistically significant difference between the sample mean and the hypothesized population mean.")
else:
    print(f"\nSince p-value ({p_value:.4f}) >= alpha ({alpha}), we fail to reject the null hypothesis.")
    print("There is no statistically significant difference between the sample mean and the hypothesized population mean.")
```
